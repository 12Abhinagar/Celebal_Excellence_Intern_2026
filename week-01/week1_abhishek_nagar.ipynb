**Your Geometric Explanation (4.3):**

When a matrix multiplies a vector, it typically transforms the vector by both changing its direction and magnitude (length). However, an **eigenvector** is a special type of vector whose *direction* remains unchanged after being multiplied by the matrix. It only gets scaled by a scalar factor. This scalar factor is called the **eigenvalue**.

Geometrically, for a matrix to "stretch" a vector means that the vector is an eigenvector, and its corresponding eigenvalue is greater than 1 (or less than -1, which implies both stretching and reversing direction). If the eigenvalue is between 0 and 1 (or -1 and 0), the eigenvector is "squished" or shortened. An eigenvalue of 1 means no change in magnitude, and an eigenvalue of 0 means the vector is squished to the zero vector. Essentially, eigenvectors are the inherent directions along which a linear transformation acts merely by stretching or compressing, without rotation.

### 4.4 SVD & Dimensionality Reduction

1. Create a random 4×3 matrix `X` (use `np.random.seed(42)`).
2. Compute its **SVD**: `U, S, Vt = np.linalg.svd(X, full_matrices=False)`.
3. **Reconstruct** `X` from the SVD and verify it matches the original.
4. Perform a **rank-1 approximation** using only the largest singular value/vector — call it `X_approx`.
5. Explain the relationship between SVD components and PCA in the markdown cell below.

````python
np.random.seed(42)
X = np.random.randn(4, 3)

U, S, Vt = np.linalg.svd(X, full_matrices=False)  # YOUR CODE HERE

X_reconstructed = U @ np.diag(S) @ Vt  # YOUR CODE HERE — reconstruct from U, S, Vt
X_approx        = U[:, :1] @ np.diag(S[:1]) @ Vt[:1, :]  # YOUR CODE HERE — rank-1 approximation

print(f"U shape: {U.shape}, S shape: {S.shape}, Vt shape: {Vt.shape}")
print(f"Singular values: {S}")
print(f"Reconstruction error: {np.linalg.norm(X - X_reconstructed):.2e}")
print(f"Rank-1 approximation:\n{X_approx}")

assert np.allclose(X, X_reconstructed, atol=1e-10), "Reconstruction failed!"
print("4.4 passed ✓")
````

**SVD → PCA Connection (4.4):**

In Singular Value Decomposition (SVD), for a matrix X:
- **U** is the matrix of left singular vectors (columns are orthonormal eigenvectors of X @ X.T).
- **S** is a diagonal matrix (or vector) containing the singular values, which are the square roots of the eigenvalues of X.T @ X (or X @ X.T) in descending order. These values represent the strength or importance of each singular vector/principal component.
- **Vt** (V transpose) is the matrix of right singular vectors (rows are orthonormal eigenvectors of X.T @ X).

**Relationship to PCA:** When PCA is applied to a data matrix X (where rows are observations and columns are features), the principal components are precisely the right singular vectors (the rows of Vt, or columns of V). The singular values (in S) indicate the variance captured by each principal component. The first singular vector (first column of V, or first row of Vt) corresponds to the first principal component, which captures the direction of maximum variance in the data.

Therefore, the **Vt** matrix (specifically its rows, which are the columns of V) contains the directions of maximum variance, representing the principal components.

### 5.4 Distribution Testing & Stationarity

1. Generate two samples: `s1` ~ Normal(0, 1), `s2` ~ Exponential(1), both n=200 (seed=0).
2. Run the **Kolmogorov-Smirnov test** (`stats.kstest`) comparing each against a standard normal CDF.
3. Generate a non-stationary time series (linear trend + noise). Run the **Augmented Dickey-Fuller test** (`adfuller` from `statsmodels`) and interpret the result.
4. Difference the series once and re-run ADF to confirm stationarity.

````python
from statsmodels.tsa.stattools import adfuller

np.random.seed(0)
s1 = np.random.normal(0, 1, 200)
s2 = np.random.exponential(1, 200)

# KS test — compare each against N(0,1)
ks_s1 = stats.kstest(s1, 'norm', args=(0, 1))  # YOUR CODE HERE
ks_s2 = stats.kstest(s2, 'norm', args=(0, 1))  # YOUR CODE HERE
print(f"KS test s1 (normal):      stat={ks_s1.statistic:.4f}, p={ks_s1.pvalue:.4f}")
print(f"KS test s2 (exponential): stat={ks_s2.statistic:.4f}, p={ks_s2.pvalue:.4f}")

# ADF test on non-stationary series
t = np.arange(200)
ts = 0.05 * t + np.random.normal(0, 1, 200)

adf_result = adfuller(ts)  # YOUR CODE HERE — adfuller(ts)
print(f"\nADF statistic: {adf_result[0]:.4f}")
print(f"ADF p-value:   {adf_result[1]:.4f}")
print(f"Stationary?    {adf_result[1] < 0.05}")

# Difference once and re-run
ts_diff = np.diff(ts)  # YOUR CODE HERE — np.diff
adf_diff = adfuller(ts_diff)  # YOUR CODE HERE
print(f"\nAfter differencing — p-value: {adf_diff[1]:.4f}, Stationary? {adf_diff[1] < 0.05}")
````

````python
def compute_psi(expected, actual, bins=10):
    """
    PSI = sum((actual_pct - expected_pct) * ln(actual_pct / expected_pct))
    Add epsilon=1e-10 to avoid log(0).
    """
    # Create bins
    min_val = min(np.min(expected), np.min(actual))
    max_val = max(np.max(expected), np.max(actual))
    bins = np.linspace(min_val, max_val, bins + 1)

    # Calculate counts for each bin
    expected_counts, _ = np.histogram(expected, bins=bins)
    actual_counts, _ = np.histogram(actual, bins=bins)

    # Convert counts to percentages (avoid division by zero)
    expected_pct = expected_counts / len(expected)
    actual_pct = actual_counts / len(actual)

    # Add a small epsilon to handle zero percentages
    epsilon = 1e-10
    expected_pct = np.maximum(expected_pct, epsilon)
    actual_pct = np.maximum(actual_pct, epsilon)

    # Calculate PSI
    psi_value = np.sum((actual_pct - expected_pct) * np.log(actual_pct / expected_pct))
    return psi_value

np.random.seed(1)
train_dist = np.random.normal(50, 10, 1000)    # training distribution
drift_dist = np.random.normal(65, 12, 1000)    # shifted production distribution

psi_value = compute_psi(train_dist, drift_dist)
print(f"PSI: {psi_value:.4f}")
print(f"Shift severity: {'Major' if psi_value > 0.2 else 'Minor' if psi_value > 0.1 else 'Stable'}")

# Plot both distributions
fig, ax = plt.subplots()
ax.hist(train_dist, bins=30, alpha=0.5, label='Training Distribution', density=True)
ax.hist(drift_dist, bins=30, alpha=0.5, label='Production Distribution', density=True)
ax.legend(); ax.set_title('Distribution Shift Visualization')
plt.tight_layout(); plt.show()
````

**Concept Drift vs Covariate Drift & PSI Thresholds (5.5):**

- **Concept drift:** Occurs when the relationship between the input variables and the target variable changes over time. This means that the underlying 'concept' the model is trying to learn has evolved.
- **Covariate drift:** Occurs when the distribution of the input variables (features) changes over time, but the relationship between the inputs and the target variable remains the same. The model sees new patterns in the input data it hasn't been trained on.
- **PSI < 0.1:** Indicates no significant shift in distribution (stable).
- **PSI 0.1–0.2:** Indicates a minor shift in distribution, which warrants investigation.
- **PSI > 0.2:** Indicates a major shift in distribution, suggesting that the model's performance may be significantly degraded.
- **Retraining trigger example:** A sudden and sustained increase in the PSI value for a key feature (e.g., above 0.2) over a period of consecutive days, indicating a major covariate drift that is likely impacting model performance.

### 6.3 Bayes' Theorem

**Scenario — Spam filter:**
- P(Spam) = 0.30 (prior)
- P("free" | Spam) = 0.80 (likelihood)
- P("free" | Not Spam) = 0.05

1. Compute P(Spam | "free") using Bayes' theorem.
2. Implement a simple `naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham)` that returns the posterior P(Spam | word).
3. In the markdown cell, map each term — **prior, likelihood, evidence, posterior** — to the spam example.

````python
p_spam            = 0.30
p_free_given_spam = 0.80
p_free_given_ham  = 0.05
p_ham             = 1 - p_spam

# P(free) — law of total probability
p_free = (p_free_given_spam * p_spam) + (p_free_given_ham * p_ham)  # YOUR CODE HERE

# Bayes
p_spam_given_free = (p_free_given_spam * p_spam) / p_free  # YOUR CODE HERE

print(f"P('free'): {p_free:.4f}")
print(f"P(Spam | 'free'): {p_spam_given_free:.4f}")

def naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham):
    # YOUR CODE HERE
    p_ham = 1 - prior_spam
    p_word = (p_word_given_spam * prior_spam) + (p_word_given_ham * p_ham)
    if p_word == 0:
        return 0 # Avoid division by zero
    return (p_word_given_spam * prior_spam) / p_word

pred = naive_bayes_predict(0.30, 0.80, 0.05)
assert abs(pred - p_spam_given_free) < 1e-9
print(f"naive_bayes_predict: {pred:.4f}")
print("6.3 passed ✓")
````

**Bayes Term Mapping (6.3):**

| Term | Formula | In this example |
|------|---------|-----------------|
| Prior | P(Spam) | The initial probability that any given email is Spam (0.30) |
| Likelihood | P(word \| Spam) | The probability of seeing the word "free" given that the email is Spam (0.80) |
| Evidence | P(word) | The overall probability of seeing the word "free" in any email |
| Posterior | P(Spam \| word) | The updated probability that an email is Spam, given that it contains the word "free" |

### 6.4 Central Limit Theorem

1. Define a population following an **exponential distribution** (λ=1, size=100,000).
2. Repeatedly draw samples of size `n=30` and record their means — do this 5,000 times.
3. Plot the histogram of sample means.
4. Overlay the **theoretical normal distribution** predicted by the CLT: μ = population mean, σ = population_std / sqrt(n).
5. Run a KS test to confirm the distribution of sample means is approximately normal.

````python
np.random.seed(7)
population = np.random.exponential(scale=1.0, size=100_000)

n_samples = 5000
sample_size = 30
sample_means = np.array([np.random.choice(population, size=sample_size, replace=False).mean() for _ in range(n_samples)])  # YOUR CODE HERE — array of 5000 sample means

pop_mean = population.mean()
pop_std  = population.std()
clt_std  = pop_std / np.sqrt(sample_size)  # YOUR CODE HERE — predicted std of sample means

print(f"Population mean: {pop_mean:.4f}, Population std: {pop_std:.4f}")
print(f"Sample means mean: {sample_means.mean():.4f}, std: {sample_means.std():.4f}")
print(f"CLT predicted std: {clt_std:.4f}")

# Plot
fig, ax = plt.subplots()
# YOUR CODE HERE — histogram of sample means + theoretical normal overlay
ax.hist(sample_means, bins=50, density=True, alpha=0.6, label='Sample Means Histogram')

x_norm = np.linspace(ax.get_xlim()[0], ax.get_xlim()[1], 100)
ax.plot(x_norm, stats.norm.pdf(x_norm, loc=pop_mean, scale=clt_std), 'r-', label='Theoretical Normal PDF')

ax.legend()
ax.set_xlabel('Sample Mean')
ax.set_ylabel('Density')
ax.set_title('CLT: Distribution of Sample Means')
plt.tight_layout(); plt.show()

# KS test
ks_result = stats.kstest((sample_means - pop_mean) / clt_std, 'norm') # Standardize for kstest against N(0,1)
print(f"KS test p-value: {ks_result.pvalue:.4f} → Approximately normal? {ks_result.pvalue > 0.05}")
````

**CLT Reflection (6.4):**

> *In 2–3 sentences: Why does this result matter for statistics and machine learning? What assumption does the CLT justify in many classical statistical tests?*

The Central Limit Theorem is crucial because it states that the distribution of sample means will be approximately normal, regardless of the population's distribution, as long as the sample size is sufficiently large. This allows us to use parametric statistical tests that assume normality (like t-tests) even when dealing with non-normally distributed data, by applying them to sample means. In machine learning, it justifies the use of certain estimation techniques and confidence intervals, simplifying the analysis of model performance and parameter estimation.
